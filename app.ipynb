```python
from flask import Flask, render_template, request
import numpy as np
import pickle

app = Flask(__name__)

# Load model & scaler
model = pickle.load(open('model.pkl', 'rb'))
scaler = pickle.load(open('scaler.pkl', 'rb'))

@app.route('/')
def home():
    return render_template('index.html')


@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Input features (correct order)
        features = [
            float(request.form['gre']),
            float(request.form['toefl']),
            float(request.form['rating']),
            float(request.form['sop']),
            float(request.form['lor']),
            float(request.form['cgpa']),
            float(request.form['research'])
        ]


        # Prediction
        prediction = model.predict([features])[0]
        prediction = round(prediction * 100, 2)

        return render_template(
            'index.html',
            prediction_text=f"Chance of getting into UCLA is: {prediction}%"
        )

    except Exception as e:
        return render_template(
            'index.html',
            prediction_text=f"Error: {str(e)}"
        )


if __name__ == "__main__":
    app.run(debug=True, use_reloader=False)
```
